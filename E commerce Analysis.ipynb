**Joining Tables**

```python
customers.head()
```

```python
geo.head()
```

The geolocation table have multiple latitude and longitude for one zip_code_prefix, before joining these tables we have to grouby the data with Average latitude and Average longitude

```python
query = '''
SELECT geolocation_zip_code_prefix, AVG(latitude) AS avg_lat, AVG(longitude) AS avg_lng, geolocation_state
FROM geo 
GROUP BY geolocation_zip_code_prefix, geolocation_state
'''

geo = pd.read_sql_query(query, con=engine)
geo.shape
```

```python
customers = customers.merge(geo, how = 'left', left_on = 'customer_zip_code_prefix', right_on = 'geolocation_zip_code_prefix')
```

```python
customers.head()
```

```python
customers.info()
```

```python
customers.drop(columns = ['geolocation_zip_code_prefix', 'geolocation_state'], inplace = True)
```

```python
customers.head()
```

```python
customers.isnull().sum()
```

We have some null values now, these are because some zip_code_prefix in customer table dont exist in geolocation table.
so lets fill these null values by mean after grouping.

```python
customers['avg_lat'] = customers.groupby('customer_city')['avg_lat'].transform(lambda x: x.fillna(x.mean()))
customers['avg_lng'] = customers.groupby('customer_city')['avg_lng'].transform(lambda x: x.fillna(x.mean()))
```

```python
customers.isnull().sum()
```

We still have some null values because those cities have not coordinate data at all.
lets fill it by mean after grouping based on state

```python
customers['avg_lat'] = customers.groupby('customer_state')['avg_lat'].transform(lambda x: x.fillna(x.mean()))
customers['avg_lng'] = customers.groupby('customer_state')['avg_lng'].transform(lambda x: x.fillna(x.mean()))
```

```python
customers.isnull().sum()
```

Now the data is cleaned and ready to use.

Joining the tables with items table to make a sales table.

```python
query = '''
SELECT i.order_id, i.product_id, i.order_item_id, i.seller_id, i.price, i.freight_value,
o.customer_id, o.order_purchase_timestamp, o.order_delivered_customer_date, o.order_status, o.is_delayed,
r.review_type,
p.product_category_name, p.product_weight_g,
s.seller_zip_code_prefix, s.seller_city, s.seller_state,
pay.total_payment
FROM items AS i
LEFT JOIN orders AS o ON
i.order_id = o.order_id
LEFT JOIN products AS p ON
i.product_id = p.product_id
LEFT JOIN reviews AS r ON
i.order_id = r.order_id
LEFT JOIN sellers AS s ON
i.seller_id = s.seller_id
LEFT JOIN payments AS pay ON
i.order_id = pay.order_id
'''

sales = pd.read_sql_query(query, engine)
```

```python
sales.head()
```

```python
sales.info()
```

```python
sales.isnull().sum()
```

Null values in date columns shows that the orders are canceled, invoiced, unavailable, or shipped, If we fill these null values with random date or if we delete these rows it would affect analysis.
Some null values in review type shows that not every customers leave the rating for products, I will replace them with 'No Rating'.
Null values in product category name shows that some products that are ordered not exists in the products table, we will fill them with 'unknown', if i drop them it will affect the total revenue.
Total payment is just calculation of price + freight_value.

```python
sales['review_type'] = sales['review_type'].fillna('No Rating')
sales['product_category_name'] = sales['product_category_name'].fillna('Unknown')
sales['total_payment'] = sales['total_payment'].fillna(sales['price'] + sales['freight_value'])
```

```python
sales.isnull().sum()
```

```python
sales.duplicated().sum()
```

Now there are some duplicates values in the dataset, dropping them does not affect the sales.

```python
sales.drop_duplicates(inplace = True)
```
